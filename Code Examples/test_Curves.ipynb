Testing OCPC module: implement a Principal Curve and plot a curve

```python
from ocpc_py import Kseg
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

Opening the synthetic dataset and get the inlier class for modeling the Curve

```python
df = sio.loadmat('flame_data.mat')
df_x = df['X']
df_y = df['y']
X = np.concatenate((df_x, df_y), axis = 1)
test_data = X[0:1000, 0:-1]
```

Fit the Principal Curve

```python
pc = Kseg(f=0.4)
pc.fitCurve(test_data)
```

Plot the data and the generated Curve

```python
fig, ax = plt.subplots()
ax.plot(test_data[:,0], test_data[:,1], 'o', label = 'data')
pc.plot_curve(ax)
ax.legend()
```

Calculate the distance between the data and the Curve

```python
_, d = pc.map_to_arcl(test_data)
print(np.mean(d))
```
